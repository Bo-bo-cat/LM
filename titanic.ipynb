# Imports

```python
import sklearn
import pandas as pd
import missingno as msno
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn import preprocessing
from sklearn.decomposition import PCA
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split as split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df["Survived"].value_counts()
```

```python
le = LabelEncoder()
df['Cabin']=df['Cabin'].str[0]
df['Cabin']=le.fit_transform(df['Cabin'])
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
 
las = LabelEncoder()
df['Sex']=las.fit_transform(df['Sex'])

df["Survived"]=df["Survived"].astype("str")


```

```python
df_train ,df_test = split(df, test_size=0.2, shuffle=True, random_state=45)
```

```python
df_train
```

```python
df_test
```

```python
train_collumns = [ "SibSp", "Parch", "Fare", 'Cabin_1', 'Cabin_2', 'Cabin_3', 'Cabin_4', 'Cabin_5', 'Cabin_6', 'Cabin_7' , 'Cabin_8', 'Sex', 'Embarked']
```

```python
label_collumn="Survived"
```

```python
x_train = df_train[train_collumns]
y_train = df_train[label_collumn] 
```

```python
x_test = df_test[train_collumns]
y_test = df_test[label_collumn] 
```

```python
model=SVC()
```

```python
model.fit(x_train, y_train)
```

```python
predictions=model.predict(x_test)
```

```python
print (classification_report(y_test, predictions))
```

```python
confusion_matrix(y_test, predictions)
```
